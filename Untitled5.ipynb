```python
import re

url = "https://ru.tradingview.com/symbols/ONDOUSD/"
text = '<span>1 год</span> <span class="change-tEo1hPMj">64,38%</span>' #вручную взят с кода сайта
pattern = r'<span>(\d+) год</span>.*?<span[^>]*>([\d,]+)%</span>'#[^>]*> подправлено при помощи ии, так как до этого не было закрывающей >, что приводило к тому, что тег не закрывался

match = re.search(pattern, text)
if match:
    year = match.group(1)
    percent = match.group(2)
    result = {"year": year,"percent": percent}
    print(f"Рост ONDO (Год: {result['year']}, Процент: + {result['percent']}%)")
```

```python
#использую не в гугл-коллаб, а в vs-code

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
import time
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.get("https://ondo.finance/")

text_to_number = {
    "Supported Chains": "10 Supported Chains",
    "Integrated Projects": "50 Integrated Projects",
    "Yieldcoin Market Share1": "80% Yieldcoin Market Share",
    "TVL": "985M TVL"
}

try:
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(15)


    data = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".MuiTypography-root.MuiTypography-h6-az.css-15w3lsz"))
    )
    for element in data:
        text = element.text
        if text in text_to_number:
            text = text_to_number[text]
            print(text)
finally:
    driver.quit()
```
